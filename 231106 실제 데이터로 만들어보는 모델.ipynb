```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

#데이터를 불러 옵니다.
df = pd.read_csv("/content/drive/MyDrive/house_train.csv")

# 데이터를 미리 살펴 보겠습니다.
print("Original DataFrame:")
print(df.head())

# 데이터 전체보기
pd.set_option('display.max_rows', None, 'display.max_columns', None)

print(df.isnull().sum().sort_values(ascending=False).head(50))

# 원핫인코딩
pre_df = pd.get_dummies(df)
print("DataFrame with One-Hot Encoding:")
print(pre_df.head())

# 전처리(빈값 채우기)
fill_pre_df = pre_df.fillna(df.mean(numeric_only=True))

# 어떻게 채워졌는지 확인
print("DataFrame with NaN Filled:")
print(fill_pre_df.head())
```
